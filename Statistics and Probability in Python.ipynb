```python
#python Mean
import numpy as np
data= np.array([12, 15, 20, 20, 22, 25])
np.mean(data)
```

```python
#python Median
import numpy as np
data= np.array([12, 15, 20, 20, 22, 25])
np.median(data)
```

```python
#varience  deviation
import numpy as np

data = [12, 15, 20, 20, 22, 25]

print("Variance:", np.var(data))
```

```python
#Standard deviation
import numpy as np
data = [12, 15, 20, 20, 22, 25]
print("Standard Deviation:", np.std(data))
```

```python
#uniform distribution
water = np.random.randint(1,20,700)
sns.histplot(water,bins=6,kde=False)
plt.title("Uniform Distribution")
plt.show()
```

```python
#Binomial Distribution
n =10 # number of trial
p =0.5 # probility of success
data = np.random.binomial(n=n,p=p,size=1000)
sns.histplot(data,kde=True)
plt.title("Binomial Distribution(cion flips)")
plt.show()
```

```python
#Normal Distribution
data = np.random.normal(loc=50,scale=10,size=500)
sns.histplot(data,kde=True)
plt.title("Normal Distribution")
plt.show()
```

```python
import numpy as np
from scipy import stats

scores = np.array([72, 68, 75, 70, 74])

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(scores, 70)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
